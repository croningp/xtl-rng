```python
from  mt19937predictor import MT19937Predictor 
import random, time, os, sys
import base64
from Crypto.Cipher import AES
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline

#in order to find where the tests are located
testing_dir = os.path.join('../randomness_testing/sp800_22_tests')
sys.path.append(testing_dir)

# determine if we are using Windows or Linux filing conventions and assign root folders
if sys.platform == 'win32':
    SCAPA_ROOT = os.path.join('Z:', os.sep, 'group')
    ORKNEY_ROOT = os.path.join('U:', os.sep)
else:
    SCAPA_ROOT = '/mnt/scapa4'
    ORKNEY_ROOT  ='/mnt/orkney1'
ORKNEY_TEAM = os.path.join(ORKNEY_ROOT, 'Clusters')
ORKNEY_PROJECT = os.path.join(ORKNEY_TEAM, 'RandomXtl')
ORKNEY_MASKRCNN = os.path.join(ORKNEY_PROJECT, 'MaskRCNN')
ORKNEY_DETECTED = os.path.join(ORKNEY_MASKRCNN, 'detected')
ORKNEY_TRAINING = os.path.join(ORKNEY_MASKRCNN, 'training')
ORKNEY_DATASETS = os.path.join(ORKNEY_TRAINING, 'datasets')
ORKNEY_LOGS = os.path.join(ORKNEY_TRAINING, 'logs')

ORKNEY_IMAGES = os.path.join(ORKNEY_PROJECT, 'images')
ORKNEY_RAW_IMGS = os.path.join(ORKNEY_IMAGES, 'raw_images')
ORKNEY_PART_IMGS = os.path.join(ORKNEY_IMAGES, 'partitioned')

ORKNEY_CV = os.path.join(ORKNEY_PROJECT, 'computer_vision')
ORKNEY_VIALS = os.path.join(ORKNEY_CV, 'vials')

ORKNEY_RANDOM = os.path.join(ORKNEY_PROJECT, 'random numbers')

def lin2win(filepath):
    return filepath.replace('/mnt/scapa4', 'Z:').replace('/mnt/orkney1', 'U:').replace('/', '\\')

def win2lin(filepath):
    return filepath.replace('Z:', '/mnt/scapa4' ).replace('U:', '/mnt/orkney1').replace('\\', '/')

def load_string(input_path):

    with open(input_path,'r') as f:
        string = f.read()
    return string

def binary2decimal(bitlist):
    if type(bitlist) == list:
        bitlist = ''.join([str(i) for i in bitlist])
    return int(bitlist, 2)

def decimal2binary(decimal, length):
    return bin(decimal)[2:].zfill(length)
    
def list2string(bitlist):
    return ''.join([str(i) for i in bitlist])

def string2list(bitstring):
    return [int(i) for i in bitstring]
    
def getcrackerstate(cracker):
    return [binary2decimal(i) for i in cracker.mt]

def get_keys_from_string(string, key_length, number_of_keys, as_list=True):
    keys = []
    for keyidx in range(number_of_keys):
        key = string[keyidx*key_length:(keyidx+1)*key_length]
        if as_list:
            key = [int(i) for i in key]           
        keys.append(key)
    return keys

def get_keys_from_MT(key_length, keys_desired, as_list=True):
    MT_keys_to_request = int(keys_desired*key_length/32)
    keys_per_request = int(32/key_length)
    keys = []
    for i in range(MT_keys_to_request):
        randint = random.getrandbits(32)
        randbits = decimal2binary(randint, 32)
        for keyidx in range(keys_per_request):
            key = randbits[keyidx*key_length:(keyidx+1)*key_length]
            if as_list:
                key = [int(i) for i in key]
            keys.append(key)
    return keys
        
    

Cu_path = r'U:\Clusters\RandomXtl\random numbers\microstates mod\CuSO4\180907d\reaction_15\032\strings\all.txt'
W_path = r'U:\Clusters\RandomXtl\random numbers\microstates mod\W19\20180214-0\Reaction_009\Image_021\strings\all.txt'
Co_path = r'U:\Clusters\RandomXtl\random numbers\microstates mod\Co4\20180216-0\Reaction_031\Image_071\strings\all.txt'

# Cu_path = win2lin(Cu_path)
# W_path = win2lin(W_path)
# Co_path = win2lin(Co_path)


Cu_string = load_string(Cu_path)
W_string = load_string(W_path)
Co_string = load_string(Co_path)

MT_state_size = 624
MT_bit_size = 32


MT_total_bits = MT_state_size * MT_bit_size # = 19968

```
